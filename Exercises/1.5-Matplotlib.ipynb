```python
import matplotlib.pyplot as plt
```

```python
lista = [1, 2, 5, 7, 3, 1]

plt.plot(lista)
plt.show()
#Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical, y 
# utilizará el índice de cada punto de datos en el array como la coordenada horizontal.
```

```python
#También se puede proporcionar dos matrices: una para el eje horizontal, y otra para el eje vertical.
lista = [-3, -1, 0, 4, 7] 
lista2 = [1, 4, 6, 7, 8]

plt.plot(lista, lista2)
plt.show()
```

```python
# Pueden modificarse las logitudes de los ejes para que la figura no se vea tan ajustada
lista = [-3, -1, 0, 4, 7] 
lista2 = [1, 4, 6, 7, 8]

plt.plot(lista, lista2)
plt.axis([-4, 8, 0, 10])
plt.show()
```

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True) #Agrega la cuadricula
plt.show()
```

```python
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

```python
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt.show()
# Esto es lo mismo que lo anterior, los plot se acumulan en la misma grafica. 
```

```python
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', label="x**2")
plt.plot(x, y2, 'g', label="x+1")
plt.legend(loc="best") # Situa el cuadro en la mejor localización
plt.show()
```

```python
# Tambien puede crearse dos graficas que no se superpongan. Estas graficas se organizan en un grid
# y se denominan subplots.
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1) # 1 rows, 1 columns, 1st subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 rows, 2 columns, 2nd subplot
plt.plot(x, y2, 'g')

plt.show()
```

```python
# Creación de figuras más grandes y con más información
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1) # 1 rows, 1 columns, 1st subplot
plt.plot(x, y, 'b--')
plt.xlabel("x1", fontsize=14)
plt.ylabel("y1", fontsize=14)

plt.subplot(1, 2, 2) # 1 rows, 2 columns, 2nd subplot
plt.plot(x, y2, 'g')
plt.xlabel("x2", fontsize=14)
plt.ylabel("y2", fontsize=14)

plt.show()
```

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211) # 2 rows, 1 columns, 1st subplot
plt.hist(data, bins = 10, rwidth=0.8) # Cantidad de barras, ancho de las barras

plt.subplot(212) # 2 rows, 1 columns, 2nd subplot
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth=0.95)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
